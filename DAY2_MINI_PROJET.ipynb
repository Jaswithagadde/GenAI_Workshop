```python
from mistralai.client import MistralClient
from mistralai.models.chat_completion import ChatMessage
from google.colab import userdata
```

```python
API_KEY = userdata.get('MISTRAL_API_KEY')

client = MistralClient(api_key=API_KEY)
```

```python
def prompt_expander(user_prompt):

    response = client.chat(
        model="mistral-small-latest",
        temperature=0.9,
        messages=[
            ChatMessage(
                role="system",
                content="you are a prompt expander and good understander of prompt"
            ),
            ChatMessage(
                role="user",
                content=user_prompt
            )
        ]
    )

    return response.choices[0].message.content



def generate_aiquote_response(user_prompt):

    response = client.chat(
        model="mistral-small-latest",
        messages=[
            ChatMessage(
                role="system",
                content="you are a quote generator"
            ),
            ChatMessage(
                role="user",
                content=expanded_prompt
            )
        ]
    )
    expanded_prompt=prompt_expander(user_prompt)
    return response.choices[0].message.content


def generate_aistory_response(user_prompt):

    response = client.chat(
        model="mistral-small-latest",
        temperature=0.9,
        messages=[
            ChatMessage(
                role="system",
                content="you are a story generator"
            ),
            ChatMessage(
                role="user",
                content=expanded_prompt
            )
        ]
    )
    expanded_prompt=prompt_expander(user_prompt)
    return response.choices[0].message.content



def generate_aiemail_response(user_prompt):

    response = client.chat(
        model="mistral-small-latest",
        messages=[
            ChatMessage(
                role="system",
                content="you are an email generator"
            ),
            ChatMessage(
                role="user",
                content=expanded_prompt
            )
        ]
    )
    expanded_prompt=prompt_expander(user_prompt)
    return response.choices[0].message.content


def ai_resume_summariser(user_prompt):

    response = client.chat(
        model="mistral-small-latest",
        temperature=0.9,
        messages=[
            ChatMessage(
                role="system",
                content="you are a AI resume summariser"
            ),
            ChatMessage(
                role="user",
                content=expanded_prompt
            )
        ]
    )

    expanded_prompt=prompt_expander(user_prompt)
    return response.choices[0].message.content


def ai_blog_into_generator(user_prompt):

    response = client.chat(
        model="mistral-small-latest",
        temperature=0.9,
        messages=[
            ChatMessage(
                role="system",
                content="you are a blog post generator"
            ),
            ChatMessage(
                role="user",
                content=expanded_prompt
            )
        ]
    )

    expanded_prompt=prompt_expander(user_prompt)
    return response.choices[0].message.content

def ai_product_description_generator(user_prompt):

    response = client.chat(
        model="mistral-small-latest",
        temperature=0.9,
        messages=[
            ChatMessage(
                role="system",
                content="you are a product description generator"
            ),
            ChatMessage(
                role="user",
                content=expanded_prompt
            )
        ]
    )
    expanded_prompt=prompt_expander(user_prompt)
    return response.choices[0].message.content


def ai_social_media_post_generator(user_prompt):

    response = client.chat(
        model="mistral-small-latest",
        temperature=0.9,
        messages=[
            ChatMessage(
                role="system",
                content="you are a social media post generator"
            ),
            ChatMessage(
                role="user",
                content=expanded_prompt
            )
        ]
    )
    expanded_prompt=prompt_expander(user_prompt)
    return response.choices[0].message.content

def ai_interview_question_generator(user_prompt):

    response = client.chat(
        model="mistral-small-latest",
        temperature=0.9,
        messages=[
            ChatMessage(
                role="system",
                content="you are a interview question generator and trainer"
            ),
            ChatMessage(
                role="user",
                content=expanded_prompt
            )
        ]
    )
    expanded_prompt=prompt_expander(user_prompt)
    return response.choices[0].message.content


print("enter 1 for Quote Generator")
print("enter 2 for Story Generator")
print("enter 3 for Email Generator")
print("enter 4 for Resume Generator")
print("enter 5 for Blog Intro Generator")
print("enter 6 for Product Description Generator")
print("enter 7 for Social Media Post Generator")
print("enter 8 for Interview Quetion Generator")
print("enter 9 for exit ")
option = input("Enter your option : ")

if option == '1':
  print("\n===== QUOTE GENERATOR =====\n")
  prompt = input("Enter your prompt: ")
  output = generate_aiquote_response(prompt)
  print("\n===== AI OUTPUT =====\n")
  print(output)
elif option == '2':
  print("\n===== STORY GENERATOR =====\n")
  prompt = input("Enter your prompt: ")
  output = generate_aistory_response(prompt)
  print("\n===== AI OUTPUT =====\n")
  print(output)
elif option == '3':
  print("\n===== EMAIL GENERATOR =====\n")
  prompt = input("Enter your prompt: ")
  output = generate_aiemail_response(prompt)
  print("\n===== AI OUTPUT =====\n")
  print(output)
elif option == '4':
  print("\n===== RESUME SUMMARY GENERATOR =====\n")
  name = input("Enter name: ")
  skills = input("Enter skills: ")
  prompt = f"""
    Create a professional resume summary for:
    Name: {name}
    Skills: {skills}
    Write a concise and professional resume summary in 3-4 lines.
    """
  output = ai_resume_summariser(prompt)
  print("\n===== AI OUTPUT =====\n")
  print(output)
elif option == '5':
  print("\n===== BLOG INTRO GENERATOR =====\n")
  topic = input("Enter blog topic: ")

  prompt = f"""Write an engaging blog introduction about:{topic}
  Keep it under 150 words."""
  output = ai_blog_into_generator(prompt)
  print("\n===== AI OUTPUT =====\n")
  print(output)

elif option == '6':
  print("\n===== PRODUCT DESCRIPTION GENERATOR =====\n")
  product = input("Enter product name: ")
  prompt = f"""Create a product description for:{product}
  Make it attractive and professional."""
  output = ai_product_description_generator(prompt)
  print("\n===== AI OUTPUT =====\n")
  print(output)
elif option == '7':
  print("\n===== SOCIAL MEDIA POST GENERATOR =====\n")
  topic = input("Enter topic: ")
  prompt = f"""Write a LinkedIn post about:{topic}
  Keep it professional and keep it under 100 words."""
  output = ai_social_media_post_generator(prompt)
  print("\n===== AI OUTPUT =====\n")
  print(output)
elif option == '8':
  print("\n===== INTERVIEW QUESTION GENERATOR =====\n")
  skill = input("Enter skill: ")
  prompt = f"""Generate 10 interview questions for:{skill}"""
  output = ai_interview_question_generator(prompt)
  print("\n===== AI OUTPUT =====\n")
  print(output)
elif option == '9':
  print("Exiting...")
else:
  print("Invalid option")
```
