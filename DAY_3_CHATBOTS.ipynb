```python
from mistralai.client import MistralClient
from mistralai.models.chat_completion import ChatMessage
from google.colab import userdata
```

```python
API_KEY = userdata.get('MISTRAL_API_KEY')

client = MistralClient(api_key=API_KEY)
```

Task 1 — Personal Memory Chatbot
A Personal Memory Chatbot is a chatbot that remembers information shared by the user and uses it in future conversations.

What it can remember

User's name

Favorite programming language

Career goals

Interests and hobbies

Preferred learning style

```python
# ============================================
# PERSONAL MEMORY CHATBOT
# ============================================


messages = []

print("===== PERSONAL MEMORY CHATBOT =====")
print("Type 'exit' to stop\n")

while True:

    user_input = input("Your Query")

    if user_input.lower() == "exit":
        print("Chat Ended")
        break


    messages.append(
        ChatMessage(
            role="user",
            content=user_input
        )
    )


    response = client.chat(
        model="mistral-small-latest",
        messages=messages
    )

    ai_reply = response.choices[0].message.content

    print("\nAI:", ai_reply)
    print()


    messages.append(
        ChatMessage(
            role="assistant",
            content=ai_reply
        )
    )
```

Task 2 — Study Buddy Chatbot
A Study Buddy Chatbot is a chatbot that helps students learn, understand concepts, solve doubts, and prepare for exams.

Features

Explain concepts in simple language

Answer study-related questions

Summarize topics

Provide examples and practice questions

Help with assignments and revision

```python
# ============================================
# STUDY BUDDY CHATBOT
# ============================================

messages = [
    ChatMessage(
            role="system",
            content="""
            You are a Study Buddy Chatbot.
            Explain concepts simply.
            Give examples.
            Summarize topics.
            Create practice questions.
            Help with assignments.
            """
        ),


print("===== STUDY BUDDY CHATBOT =====")
print("Type 'exit' to stop\n")

while True:

    user_input = input("Your Query")

    if user_input.lower() == "exit":
        print("Chat Ended")
        break


    messages.append(
        ChatMessage(
            role="user",
            content=user_input
        )
    )


    response = client.chat(
        model="mistral-small-latest",
        messages=messages
    )

    ai_reply = response.choices[0].message.content

    print("\nAI:", ai_reply)
    print()


    messages.append(
        ChatMessage(
            role="assistant",
            content=ai_reply
        )
    )
```

Task 3 — Multi-Personality Chatbot
A Multi-Persona Chatbot is a chatbot that can switch between different personalities, roles, or expertise areas based on the user's choice.

Example

Instead of creating a separate chatbot for each role:

Data Science Bot

Travel Bot

Fitness Bot

Career Bot

You create one chatbot that can act as all of them.

```python
# ============================================
# MULTI-PERSONALITY CHATBOT
# ============================================


messages = [
    ChatMessage(
            role="system",
            content="""
            you are a multi-persona chatbot.
            You can switch between different personalities, roles, or expertise areas based on the user's choice.
            """
        ),

]

print("===== MULTI-PERSONALITY CHATBOT =====")
print("Type 'exit' to stop\n")

while True:

    user_input = input("Your Query")

    if user_input.lower() == "exit":
        print("Chat Ended")
        break


    messages.append(
        ChatMessage(
            role="user",
            content=user_input
        )
    )


    response = client.chat(
        model="mistral-small-latest",
        messages=messages
    )

    ai_reply = response.choices[0].message.content

    print("\nAI:", ai_reply)
    print()


    messages.append(
        ChatMessage(
            role="assistant",
            content=ai_reply
        )
    )
```

Task 4 — Interview Preparation Chatbot
Objective: Ask interview questions and provide feedback.

Example:

Bot: What is Machine Learning? User: ... Bot: Good answer! Here's how you can improve it...

```python
# ============================================
# INTERVIEW PREPARATIONAND FEEDBACK CHATBOT
# ============================================

messages = [
    ChatMessage(
            role="system",
            content="""
            you are a interview preparation chatbot.
            your goal is to ask interview questions one by one and provide feedback to user answer.
            """
        ),
]

print("===== INTERVIEW PREPARATIONAND FEEDBACK CHATBOT =====")
print("Type 'exit' to stop\n")

while True:

    user_input = input("Your Query")

    if user_input.lower() == "exit":
        print("Chat Ended")
        break


    messages.append(
        ChatMessage(
            role="user",
            content=user_input
        )
    )


    response = client.chat(
        model="mistral-small-latest",
        messages=messages
    )

    ai_reply = response.choices[0].message.content

    print("\nAI:", ai_reply)
    print()


    messages.append(
        ChatMessage(
            role="assistant",
            content=ai_reply
        )
    )
```
