**1. Sentiment Analysis**

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch
```

```python
model = AutoModelForSequenceClassification.from_pretrained(
    "distilbert-base-uncased-finetuned-sst-2-english"
)

tokenizer = AutoTokenizer.from_pretrained(
    "distilbert-base-uncased-finetuned-sst-2-english"
)
```

```python
text = input("Enter text: ")
```

```python
inputs = tokenizer(text, return_tensors="pt")

print(inputs)
```

```python
outputs = model(**inputs)
print(outputs.logits)
```

```python
predictions = torch.nn.functional.softmax(outputs.logits, dim=-1)
print(predictions)
```

```python
prediction_class_id = outputs.logits.argmax().item()

print(prediction_class_id)
```

```python
print("Sentiment:", model.config.id2label[prediction_class_id])
```

**2. Summarization**

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained(
    "facebook/bart-large-cnn"
)

tokenizer = AutoTokenizer.from_pretrained(
    "facebook/bart-large-cnn"
)
```

```python
text = input("Enter article: ")
```

```python
inputs = tokenizer(text, return_tensors="pt")
```

```python
print(inputs)
```

```python
summary = model.generate(
    inputs["input_ids"],
    max_length=50
)

print(summary)
```

```python
print(tokenizer.decode(
    summary[0],
    skip_special_tokens=True
))
```

**4. Text Generation**

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
tokenizer = AutoTokenizer.from_pretrained(
    "TinyLlama/TinyLlama-1.1B-Chat-v1.0"
)
model = AutoModelForCausalLM.from_pretrained(
     "TinyLlama/TinyLlama-1.1B-Chat-v1.0")
```

```python
user_prompt = input("Enter a prompt: ")
```

```python
inputs = tokenizer(user_prompt, return_tensors="pt")

print(inputs)
```

```python
output = model.generate(
    **inputs,
    max_new_tokens=50
)

print(output)
```

```python
response = tokenizer.decode(
    output[0],
    skip_special_tokens=True
)

print("\nGenerated Response:")
print(response)
```

5. Question Answering

```python
from transformers import AutoTokenizer,AutoModelForQuestionAnswering
```

```python
tokenizer = AutoTokenizer.from_pretrained(
    "bert-large-uncased-whole-word-masking-finetuned-squad"
)
model = AutoModelForQuestionAnswering.from_pretrained(
     "bert-large-uncased-whole-word-masking-finetuned-squad")
```

```python
context = """AIML (Artificial Intelligence Markup Language) is an XML dialect for creating natural language software agents, or chat-bots. AIML was developed by Richard Wallace and the global free software community between 1995 and 2000. It forms the basis for A.L.I.C.E. (Artificial Linguistic Internet Computer Entity), a natural language processing chatbot. AIML provides a simple way to define patterns that map user inputs to bot responses. For example, a pattern could be 'WHAT IS YOUR NAME' and the corresponding template would be 'My name is A.L.I.C.E.'. It is widely used in conversational AI applications due to its simplicity and flexibility."""
question = input("Enter the question: ")
```

```python
import torch

inputs = tokenizer(
    question,
    context,
    return_tensors="pt"
)

with torch.no_grad():
    outputs = model(**inputs)

start_index = torch.argmax(outputs.start_logits)
end_index = torch.argmax(outputs.end_logits) + 1

answer = tokenizer.decode(
    inputs["input_ids"][0][start_index:end_index]
)

print("\nAnswer:")
print(answer)
```
